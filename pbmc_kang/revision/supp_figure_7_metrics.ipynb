```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sb

font = {'family' : 'Arial',
        'size'   : 14}

matplotlib.rc('font', **font)
matplotlib.rc('ytick', labelsize=14)
matplotlib.rc('xtick', labelsize=14)
matplotlib.rcParams["figure.dpi"] = 140
```

```python
dict_0_1 = pd.read_pickle("scIB_IE_X_cvae_B_condition_merged_CT_cell_type_joint_uid1659540206.5109847.pkl")
dict_0_25 = pd.read_pickle("scIB_IE_X_cvae_B_condition_merged_CT_cell_type_joint_uid1659539872.9099467.pkl")
dict_0_5 = pd.read_pickle("scIB_IE_X_cvae_B_condition_merged_CT_cell_type_joint_uid1660059675.4730496.pkl")

df = pd.DataFrame(dict_0_1, index=["Downsample to 0.1"])
df = df.append(pd.DataFrame(dict_0_25, index=["Downsample to 0.25"]))
df = df.append(pd.DataFrame(dict_0_5, index=["Downsample to 0.5"]))
```

```python
del df["adata_name"]
```

```python
df["overall"] = df.mean(axis=1)
```

```python
df
```

```python
sb.color_palette('Set2', 3)
```

```python
axs = df.T.plot.bar(figsize=(9,7), color=sb.color_palette('Set2', 3))
```

```python
axs.set_xticklabels(axs.get_xticklabels(), rotation=45)
axs.figure
```

```python
df[["overall"]].T.plot.bar(figsize=(9,7), color=sb.color_palette('Set2', 3), ylim=(0,1))
```
